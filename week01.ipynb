### Magic Commands

* start with the `%` symbol (line magic), or with `%%` (cell magic).

````python
%timeit L = [n ** 2 for n in range(999)]
````

````python
%timeit L = [n * n for n in range(999)]
````

````python
%%timeit
L = []
for n in range(999):
    L.append(n*n)
````

````python
%timeit?
````

````python
%magic
````

````python
%lsmagic
````
